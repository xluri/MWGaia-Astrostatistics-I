## Probability Density Functions (PDF) and Cumulative Distribution functions (CDF)

* PDF: $P(a \leq x \leq b) = \int_a^b f(x) \, dx$ 

* CDF:  cdf(y) = $P(x \leq y) = \int_{-\infty}^y f(x) \, dx$

### 1- Normal distribution

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np

N = 100
min = -6.
max = 6.
x = np.linspace(min, max,N) # N points from min to max
fig = plt.figure(frameon=False)
fig.set_size_inches(15,7.5)

# ----------
# Plot PDFs 
# ----------
p1= plt.subplot(1,2,1)

mu,sig = 0.,1.
plt.plot(x, scp.norm.pdf(x,mu,sig),'r-', lw=2, label='norm')
mu,sig = 0.,2.
plt.plot(x, scp.norm.pdf(x,mu,sig),'g-', lw=2, label='norm')
mu,sig = 0.,3.
plt.plot(x, scp.norm.pdf(x,mu,sig),'b-', lw=2, label='norm')

plt.ylabel('PDF')
plt.xlabel('x')

# ----------
# Plot CDFs 
# ----------
p1= plt.subplot(1,2,2)

mu,sig = 0.,1.
plt.plot(x, scp.norm.cdf(x,mu,sig),'r-', lw=2, label='norm')
mu,sig = 0.,2.
plt.plot(x, scp.norm.cdf(x,mu,sig),'g-', lw=2, label='norm')
mu,sig = 0.,3.
plt.plot(x, scp.norm.cdf(x,mu,sig),'b-', lw=2, label='norm')

plt.ylabel('CDF')
plt.xlabel('x')

plt.show()
```

###  4 - Beta distribution

Wikipedia: "The beta distribution has been applied to model the behavior of random variables limited to intervals of finite length in a wide variety of disciplines." 

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np

N = 100
min = 0.
max = 1.
x = np.linspace(min, max,N) # N points from min to max
fig = plt.figure(frameon=False)
fig.set_size_inches(15,7.5)

# ----------
# Plot PDFs 
# ----------
p1= plt.subplot(1,2,1)

a,b = 0.5,0.5
plt.plot(x, scp.beta.pdf(x,a,b),'r-', lw=2, label='beta')
a,b = 5.,1.
plt.plot(x, scp.beta.pdf(x,a,b),'b-', lw=2, label='beta')
a,b = 1.,3.
plt.plot(x, scp.beta.pdf(x,a,b),'g-', lw=2, label='beta')
a,b = 2.,2.
plt.plot(x, scp.beta.pdf(x,a,b),'y-', lw=2, label='beta')

plt.title('PDF')
plt.xlabel('x')

# ----------
# Plot CDFs 
# ----------
p1= plt.subplot(1,2,2)

# Beta distribution
a,b = 0.5,0.5
plt.plot(x, scp.beta.cdf(x,a,b),'r-', lw=2, label='beta')
a,b = 5.,1.
plt.plot(x, scp.beta.cdf(x,a,b),'b-', lw=2, label='beta')
a,b = 1.,3.
plt.plot(x, scp.beta.cdf(x,a,b),'g-', lw=2, label='beta')
a,b = 2.,2.
plt.plot(x, scp.beta.cdf(x,a,b),'y-', lw=2, label='beta')

plt.title('CDF')
plt.xlabel('x')

plt.show()
```

### Calculating probabilities: integration vs. CDF

By definition, the proability of a random variable to be in the interval $[x_1,x_2]$ is:

$P(x_1 \leq x \leq x_2) = \int_{x_1}^{x_2} f(x) \; dx$

where $f(x)$ is the PDF. Let's implement this with a normal distribution.

```python
import scipy.stats as scp
import scipy.integrate as integrate

mu,sig = 0.,1.
Pint= integrate.quad(lambda x: scp.norm.pdf(x,mu,sig), -sig, sig)
print("P(",-sig,",",sig,")=",Pint[0],"+/-",Pint[1])
```

Now, we can repeat the calculation using the CDF. Following the definition of the CDF:

$cdf(y) = P(x \leq y) = \int_{-\infty}^y f(x) \, dx$

it follows that:

$P(x_1 \leq x \leq x_2) = \int_{x_1}^{x_2} f(x) \; dx = cdf(x_2)-cdf(x_1)$

Let's use this property:

```python
import scipy.stats as scp

mu,sig = 0.,1.
Pcdf= scp.norm.cdf(sig,mu,sig) - scp.norm.cdf(-sig,mu,sig)

print("P(",-sig,",",sig,")=",Pcdf)
```

And the difference

```python
print("Pint-Pcdf= ",Pint[0]-Pcdf)
```
