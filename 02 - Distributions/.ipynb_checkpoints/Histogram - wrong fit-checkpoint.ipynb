### IMPORTANT

This notebook is an example of a ___wrong practice___, directly comparing the PDF with an histogram 

```python
%matplotlib inline
from scipy.stats import norm
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.distributions.empirical_distribution as stm

# Parameters of the normal distribution
mu = 0
sig = 1
N = 10000

# Generate normal sample
x = np.random.normal(mu, sig,N) 

p = plt.hist(x,10,density=True)

plt.plot(x, scp.norm.pdf(x,mu,sig),'r.', lw=2, label='norm')
```
