## PDF and CDF


```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np

N = 100
min = -7.
max = 7.
x = np.linspace(min, max,N) # N points from min to max
fig = plt.figure(frameon=False)
fig.set_size_inches(15,7.5)

# ----------
# Plot PDFs 
# ----------
p1= plt.subplot(1,2,1)

mean, scale = 0., 1.
plt.plot(x, scp.cauchy.pdf(x,mean,scale),'r-', lw=2, label='cauchy')
mean, scale = 0., 2.
plt.plot(x, scp.cauchy.pdf(x,mean,scale),'g-', lw=2, label='cauchy')
mean, scale = 0., 3.
plt.plot(x, scp.cauchy.pdf(x,mean,scale),'b-', lw=2, label='cauchy')

plt.title('PDF')
plt.xlabel('x')

# ----------
# Plot CDFs 
# ----------
p1= plt.subplot(1,2,2)

mean, scale = 0., 1.
plt.plot(x, scp.cauchy.cdf(x,mean,scale),'r-', lw=2, label='cauchy')
mean, scale = 0., 2.
plt.plot(x, scp.cauchy.cdf(x,mean,scale),'g-', lw=2, label='cauchy')
mean, scale = 0., 3.
plt.plot(x, scp.cauchy.cdf(x,mean,scale),'b-', lw=2, label='cauchy')

plt.title('CDF')
plt.xlabel('x')


plt.show()
```

## Quantile function

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np

N = 100
min = 0.
max = 1.
x = np.linspace(min, max,N) # N points from min to max

# ----------
# Plot QFs 
# ----------
mean, scale = 0., 1.
plt.plot(x, scp.cauchy.ppf(x,mean,scale),'r-', lw=2, label='cauchy')
mean, scale = 0., 2.
plt.plot(x, scp.cauchy.ppf(x,mean,scale),'g-', lw=2, label='cauchy')
mean, scale = 0., 3.
plt.plot(x, scp.cauchy.ppf(x,mean,scale),'b-', lw=2, label='cauchy')

plt.title('QF')
plt.xlabel('x')

plt.show()
```

## Law of large numbers: non-convergence of mean and variance

See the section "Limitation" in the wikipedia: https://en.wikipedia.org/wiki/Law_of_large_numbers

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np

N = 1000 # Number of points for each batch generation

mean = []
median = []
variance = []
all_points = []

# Accumulate points and re-calculate statistics at each step
for i in range(200):
    all_points.extend(np.random.standard_cauchy(N))
    mean.append( np.mean(all_points) )
#     median.append( np.median(all_points) )
    variance.append( np.var(all_points) )

fig = plt.figure(frameon=False)
fig.set_size_inches(15,15)

# Plot series
p1= plt.subplot(2,2,1)
plt.plot(mean, 'r.',)
plt.title('Mean')
plt.xlabel('N')

p1= plt.subplot(2,2,2)
plt.plot(variance, 'r.',)
plt.title('Variance')
plt.xlabel('N')

# Plot histograms
p1= plt.subplot(2,2,3)
plt.hist(all_points,100)
plt.title('Full histogram')

p1= plt.subplot(2,2,4)
plt.hist(all_points,100,range=(-7,7))
plt.title('Clipped histogram')

plt.show()
```

Compare these plots with the "well behaved" distributions used to illustrate the Law of Large Numbers

## Corollary: physical interpretation

Consider a point source projecting light into a (infinite) flat screen.

You can prove that the probability of a photon hitting a given point of the screen (vertical $y$ position) follows a Cauchy distribution:

![](Cauchy2.png)

Now you can interpret the large "jumps" in the previous plots as occasional photons emitted in a quasi-vertical direction (large  $y$). The larger the sample, the more extreme these occasional photons can become, breaking the mean and the variance of the "rest" of the sample.
