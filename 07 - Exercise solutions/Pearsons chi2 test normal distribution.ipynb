### Wrong use: histogram of p-values

We apply the $\chi^2$ test as we did in the case of the uniform distribution, with equally-spaced histogram bins regardless of how many points are in each one. 

__IMPORTANT!!__  This is a counter-example. Remember than the test may not be reliable when there are few bins or when
some of the bins have less than 5 counts. 

We apply the test here without respecting these conditions and we can see how the p-value distribution is not well behaved, is not uniform.

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pylab as plt
import numpy as np

# Parameters of the uniform distributionto use
# and number of points per sample
mu= 0
sig= 4
N= 1000
Nsamples= 10000

dist= scp.norm(mu,sig)

# Generate equally spaced bins; we use the PPF to calculate cut the (infinite) tails
# of the normal distribution some very low apoints nd then equi-partition the space between
x_min= dist.ppf(0.00001)
x_max= dist.ppf(0.99999)
bins= np.linspace(x_min,x_max,20)
# print(bins)

# Predict counts from the distribution CDF. 
#
# First we calculate the prbabilities for each bin using the CDF and correct
# them to take into account that we have only covered the (x_min,x_max) 
# region, so we have to normalize the probability to one (otherwise the 
# chisquare method complains).
#
# Then we multiply by N to predict the counts.
p_bins= np.array([ (dist.cdf(bins[i+1])-dist.cdf(bins[i]) ) for i in range(len(bins)-1)] )
p_bins= p_bins / sum(p_bins)
p_bins= N*p_bins

# Plot histogram of a single sample as a reference
fig = plt.figure(frameon=False)
fig.set_size_inches(15,7.5)

p1= plt.subplot(2,1,1)
p= plt.step(bins[:-1],p_bins,where="post",color="red")
x= dist.rvs(N)
plt.xlabel('x')
plt.ylabel('Count')
p= plt.hist(x,bins)


# Generate samples and collect p-values
p_values= []
excluded= 0
for i in range(Nsamples):
    x= dist.rvs(N)
    counts= np.histogram(x,bins)
    if sum(counts[0]) != N: # Exclude samples with objects outside (x_min,x_max); otherwise the total count does not match
        ++excluded
        continue
    c2_stat, p_val = scp.chisquare(counts[0],p_bins)  
    p_values.append(p_val)
    
print(excluded," samples with objects outside (x_min,x_max) excluded")

# Plot histogram of p-values
p1= plt.subplot(2,1,2)
h= plt.hist(p_values,20) # h[0] will be the counts per bin
plt.grid(True)
plt.xlabel('p')
plt.ylabel('Count')

# Check if the distribution of p-values is normal
c2_stat, p_val = scp.chisquare(h[0]) 
print("Uniformity test of p-values: {0:.2f}%".format(100*p_val))

plt.show()
```

### Correct use: histogram of p-values

Here we define histogram bins with different sizes, aiming to have them contain enough points to fulfill the conditions of applicability of the $\chi^2$ test.

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pylab as plt
import numpy as np

# Parameters of the uniform distributionto use
# and number of points per sample
mu= 0
sig= 4
N= 1000
Nsamples= 10000

dist= scp.norm(mu,sig)

# Generate bins we create two large bins containing the tails and 
# then we equipartition the rest
x_min= dist.ppf(0.00001)
x_max= dist.ppf(0.99999)
bins=[]
bins.append(x_min) # To include the left 5% tail

for x in np.linspace(dist.ppf(0.05), dist.ppf(0.95), 18):
    bins.append(x)

bins.append(x_max) # To include the right 5% tail

# print(bins)

# Predict counts from the distribution CDF. 
#
# First we calculate the prbabilities for each bin using the CDF and correct
# them to take into account that we have only covered the (x_min,x_max) 
# region, so we have to normalize the probability to one (otherwise the 
# chisquare method complains).
#
# Then we multiply by N to predict the counts.p_bins= [ N*(dist.cdf(bins[i+1])-dist.cdf(bins[i])) for i in range(len(bins)-1)]
p_bins= np.array([ (dist.cdf(bins[i+1])-dist.cdf(bins[i]) ) for i in range(len(bins)-1)] )
p_bins= p_bins / sum(p_bins)
p_bins= N*p_bins


# Plot histogram of a single sample as a reference
fig = plt.figure(frameon=False)
fig.set_size_inches(15,7.5)

p1= plt.subplot(2,1,1)
p= plt.step(bins[:-1],p_bins,where="post",color="red")
x= dist.rvs(N)
plt.xlabel('x')
plt.ylabel('Count')
p= plt.hist(x,bins)


# Generate samples and collect p-values
p_values= []
excluded= 0
for i in range(Nsamples):
    x= dist.rvs(N)
    counts= np.histogram(x,bins)
    if sum(counts[0]) != N: # Exclude samples with objects outside (x_min,x_max); otherwise the total count does not match
        ++excluded
        continue
    c2_stat, p_val = scp.chisquare(counts[0],p_bins)  
    p_values.append(p_val)

print(excluded," samples with objects outside (x_min,x_max) excluded")

# Plot histogram of p-values
p1= plt.subplot(2,1,2)
h= plt.hist(p_values,20) # h[0] will be the counts per bin
plt.grid(True)
plt.xlabel('p')
plt.ylabel('Count')


# Check if the distribution of p-values is normal
c2_stat, p_val = scp.chisquare(h[0]) 
print("Uniformity test of p-values: {0:.2f}%".format(100*p_val))

plt.show()
```
