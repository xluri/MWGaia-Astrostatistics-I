The wikipedia entry is a good reading to introduce the ML estimator concepts:  https://en.wikipedia.org/wiki/Maximum_likelihood_estimation

In short (in data science terminology and statistics terminology)

* Given a dataset D with data for $N$ objects, where each piece of data (each realisation of a random variable) consists of $k$ measurements (multivariate random variable):

    $D = {\vec{x_1}, \ldots , \vec{x_N}}$
    
    $\vec{x_i} = (x_i^1, \ldots, x_i^k)$
    
    $\;$
* Given a physical model of these data (in other words, a PDF describing the distribution of the random variable $\vec{x}$) the probability of "observing" a given value $\vec{x}$ would be:

    $P(\vec{x}) = f(\vec{x} | \vec{\theta}) \; dx$
    
    Where $\vec{\theta} = (\theta_1, \ldots, \theta_m)$ are the "parameters of the model".
        
    $\;$
    
* Assuming that ___the observations are independent and all of them properly described by the model___, we construct the ___Likelihood function___ of the sample, given the model:

    $\mathcal{L}(D | \vec{\theta}) = \prod_{i=1}^{N} f(\vec{x_i} | \vec{\theta}) $
    
    It is related the "joint probability" of the observed values $D$ given the model $f(\vec{x} | \vec{\theta})$ (missing $dx$s).
    
    $\;$
        
* We do not know the "true" values of $\vec{\theta}$, and we want to estimate them from the dataset $D$. We define the ___Maximum Likelihood Estimator___ of these values as:

    $\hat{\vec{\theta}}$ such that  $\mathcal{L}(D | \hat{\vec{\theta}} )$ is maximum
    
    $\hat{\vec{\theta}}$ represents the ___most probable___ values of the parameters of the model given the observations in the sample $D$.
    
    $\;$
        
* In practice (for numerical reasons), it is more convenient to work with the ___log-likelihood___ function which has the same maxima:

    $\mathcal{l}(D | \vec{\theta}) = log( \mathcal{L}(D | \vec{\theta}) ) = \sum_{i=1}^{N} log( f(\vec{x_i} | \vec{\theta})) $
    
    
In some cases the maximum likelihood estimators (MLE) can be obtained analytically (see for instance its relation with the least squares method: https://bookdown.org/egarpor/PM-UC3M/app-ext-mle.html). However, usually MLE are obtained by numerical maximization of $\mathcal{l}(D | \vec{\theta})$; in this case beware of complex $\mathcal{l}$ functions that can have multiple local maxima, you are looking for the absolute maximum.

### Simple example: a univariate normal distribution

We will generate values based on a univariate normal distribution, where $\vec{\theta} = (\mu, \sigma)$:

$f(\vec{x} | \vec{\theta}) = f(x | \mu,\sigma ) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x-\mu)^2}{2 \sigma^2}}$

```python
# General imports
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.optimize as opt
import statistics as st

from scipy.stats import norm
from scipy.stats import multivariate_normal
```

```python
# Parameters of the normal distribution and number of objects in the sample
mu = 1
sigma = 2
N = 2000

# Generate numbers normally distributed. We use numpy for convenience
x = np.random.normal(mu, sigma, N)

# Generate histogram
count, bins, ignored = plt.hist(x, 30)

plt.title('Normally distributed values')
plt.ylabel('n')
plt.xlabel('value')
plt.show()
```

In _scipy_ the distributions contain a _fit()_ method that does by default a ML fit:

    fit(data, *args, **kwds)
       Return estimates of shape (if applicable), location, and scale parameters from data. 
       The default estimation method is Maximum Likelihood Estimation (MLE), but Method 
       of Moments (MM) is also available.

See the scipy documentation for more details:

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.fit.html

Let's try it for our normally distributed data:

```python
ml_mu, ml_sigma = norm.fit(x)

print("ml_mu    = ",ml_mu)
print("ml_sigma = ",ml_sigma)
```

### Simple example: repeat using explicitly a minimisation routine

We repeat the previous example but instead of using the built-in fitting routine in the _np.random.normal_ class we explicitly implement the ML maximization.

First, we define the log of the PDF function

```python
def log_f(x,theta):
    '''This function implements the PDF, a normal distribution
    * x is the measured value
    * theta is the vector of the PDF parameters: theta[0] is $\mu$ and theta[1] is $\sigma$'''
    
    return -(x-theta[0])**2 / (2*theta[1]**2) - math.log( theta[1] * math.sqrt(2*math.pi) )
```

Now we define the log-likelihood function. Notice that __the function returns minus the log-likelihood__, $-log( \mathcal{L}(D | \vec{\theta}) )$. This is because we will use a minimisation routine (this is what is provided by the python module), so by minimising $-log( \mathcal{L}(D | \vec{\theta}) )$ we are maximising the log-likelihood.

___Note:___ this implementation of the log-likelihood function is rather slow, since it using a for loop and calling the external _log\_f()_ function. It is noticeable when generating the contour plot below, which generates a grid of points. It can be made faster by intergrating the calculations of _log\_f_ and using numpy to sum the list of values instead of a for loop.

```python
def minus_log_L(theta):
    '''Minus log-likelihood function. It uses as arguments the parameters of the PDF (which will be passed by the
    minimization routine, and assumes that the vector of observed values x is defined externally as a global variable. '''

    logL= 0
    
    for value in x:
        logL= logL + log_f(value,theta)
        
    return -logL
```

Let's do a contour plot of the logL function around the "true" values of the parameters of the PDF.

```python
theta=[mu,sigma] # We set the "true" values used to generate the sample

# Grid of $\mu$ values
theta_0= [t for t in np.arange(mu-2,mu+2,0.05)]

# Grid of $\sigma$ values
theta_1= [t for t in np.arange(sigma-0.5,sigma+1,0.05)]

# Grid of values
z = [ [minus_log_L([m,t]) for m in theta_0] for t in theta_1]

# Contour plot
plt.xlabel("$\mu$")
plt.ylabel("$\sigma$")
plt.title("$-log( \mathcal{L}(D | \mu, \sigma) )$")
_= plt.contour(theta_0,theta_1,z,45)
```

We use the minimisation routine from scypy:

    https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html
    
Compare the result with the one using the _norm.fit()_ function above.

```python
# Initial guess for the values of the PDF parameters, not too far of the true values
theta_guess=[2.,3.]

# Minimization of our (minus) log-likelihood function
result= opt.minimize(minus_log_L, theta_guess, tol=1.e-4, bounds=( (None,None), (0,None) ) )
print(result)
```

See:
    
https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.OptimizeResult.html#scipy.optimize.OptimizeResult
    
for a description of the output of the minimixzer

### Exercise: bivariate uncorrelated normal distribution

We generate a two-dimensional data samplw, that is, each data point is a pair $(x_1,x_2)$. The distributions of $x_1$ and $x_2$ are normal, and the variables are independent (the covariance matrix is diagonal). Determine using ML the mean and variance in each axis.

We generate the sample here:

```python
# Means
mu= [1,5]

# Covariance matrix (diagonal, uncorrelated x1 and x2). Diagonal contains the variances, so sigma**2
cov_matrix= np.diag([4,16])

# Create distribution
mv_normal= multivariate_normal(mean=mu, cov= cov_matrix)

# Generate sample
x= mv_normal.rvs(size=2000)

# Check means and sigma
print("x1: ", st.mean(x[:,0]), st.stdev(x[:,0]))
print("x2: ", st.mean(x[:,1]), st.stdev(x[:,1]))

# Generate plot
plt.subplot(2,2,3)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.plot(x[:,0],x[:,1],".")

plt.subplot(2,2,1)
plt.hist(x[:,0])

plt.subplot(2,2,4)
plt.xlabel('$x_2$')
_ = plt.hist(x[:,1])
```

## Likelihood ratio tests

When using MLE the model used to define the Likelihood $f(\vec{x} | \vec{\theta})$ is a crucial element. Sometimes it is not clear which is the model to use, and one can always define more and more complex models, with more parameters to be determined by MLE.

In these cases the increment on the number of parameters will usually led to a better (higher likelihood) fit to the data. But is the increase of complexity worth the gain in the fit? Or are we overfitting the data?

To decide if a more complex model is required you can use a __likelihood ratio test___ as described here:

https://www.statology.org/likelihood-ratio-test-in-python/
